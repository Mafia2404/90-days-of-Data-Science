<font color="red" size=3><b>Concatenation of Data Frame</b></font>

**Note** Concatenation is not equivalent to merging

```python
df00 = pd.read_csv('wq.csv')
df00.head()
```

```python
# creating small datasets
df11 = df00[['fixed acidity', 'volatile acidity', 'citric acid']][0:5]
df11.head()
```

```python
df22 = df00[['fixed acidity', 'volatile acidity', 'citric acid']][11:15]
df22.head()
```

```python
# row wise concatenation(vertical)
pd.concat([df11, df22])
```

```python
# column wise concatenation(horizontal)
df33 = pd.concat([df11, df22],axis=1)
df33
```

<font color="red" size=3><b>Merging of Data Frames</b></font>

**Note** Merge works on columns i.e column based joining

```python
data1 = pd.DataFrame({ 's1':[1,2,45,6,67],
                      's4':[56,5,6,7,8],
                      's5':[3,56,5,6,6]
    
})
data1
```

```python
data2 = pd.DataFrame({ 's1':[1,2,3,4,5],
                      's2':[4,5,6,7,8],
                      's3':[3,56,5,6,7]},
                    )
data2
```

```python
# by default inner merge occurs (Venn diagram concept)
pd.merge(data1,data2)
```

```python
pd.merge(data1,data2, how='left')
```

```python
pd.merge(data1,data2, how='right')
```

```python
pd.merge(data1,data2, how='outer')
```

```python
pd.merge(data1,data2, how='cross')
```

<font color="red" size=3><b>Joining of Data Frames</b></font>

**Note** Joins performs index based matching

```python
data1 = pd.DataFrame({ 's1':[1,2,45,6,67],
                      's4':[56,5,6,7,8],
                      's5':[3,56,5,6,6]},
                     index=['a','b','c','d','e']
)
data1
```

```python
data2 = pd.DataFrame({ 's8':[1,2,3,4,5],
                      's2':[4,5,6,7,8],
                      's3':[3,56,5,6,7]},
                     index=['f','g','a','c','j']
    
)
data2
```

```python
data1.join(data2) # by default left join is performed
```

```python
data1.join(data2, how='inner')
```

```python
data1.join(data2, how='right')
```

```python
data1.join(data2, how='outer')
```

```python
data1.join(data2, how='cross')
```
