```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
encoder = LabelEncoder()
df['species'] = encoder.fit_transform(df['species'])
```

```python
df.head()
```

```python
df1 = df[['sepal_length','petal_length','species']]
```

```python
df2 = df[['sepal_width','petal_width','species']]
```

```python
df1.head()
```

```python
df2.head()
```

```python
X1 = df1.iloc[:,0:2]
y1 = df1.iloc[:,-1]
```

```python
X2 = df2.iloc[:,0:2]
y2 = df2.iloc[:,-1]
```

```python
X1_train,X1_test,y1_train,y1_test = train_test_split(X1,y1,test_size=0.2,random_state=2)
```

```python
X2_train,X2_test,y2_train,y2_test = train_test_split(X2,y2,test_size=0.2,random_state=2)
```

```python
clf1 = LogisticRegression(multi_class='multinomial')
```

```python
clf2 = LogisticRegression(multi_class='multinomial')
```

```python
clf1.fit(X1_train,y1_train)
```

```python
clf2.fit(X2_train,y2_train)
```

```python
y1_pred = clf1.predict(X1_test)
```

```python
y2_pred = clf2.predict(X2_test)
```

```python
print(accuracy_score(y1_test,y1_pred))
```

```python
print(accuracy_score(y2_test,y2_pred))
```

```python
pd.DataFrame(confusion_matrix(y1_test,y1_pred))
```

```python
pd.DataFrame(confusion_matrix(y2_test,y2_pred))
```
