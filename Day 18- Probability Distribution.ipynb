<font color="blue" size=5><b>Probability Distribution Function</b></font>

```python
import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
```

<font color="red" size=3><b>Probability Mass Function(pmf)</b></font>

```python
# Experiment: Rolling a dice
dice=[]
for i in range(1000):
    dice.append(random.randint(1,6))
```

```python
len(dice)
```

```python
# calculating probability 
s = pd.Series(dice).value_counts()/pd.Series(dice).value_counts().sum()
s = s.sort_index()
s
```

```python
# plotting the graph
x = s.index
y = s.values
plt.bar(x,y)
```

```python
# Experiment: Rolling 2 dice
dice1=[]
for i in range(1000):
    a = random.randint(1,6)
    b = random.randint(1,6)

    dice1.append(a+b)
```

```python
# calculating probability 
s1 = pd.Series(dice1).value_counts()/pd.Series(dice1).value_counts().sum()
s1 = s1.sort_index()
s1
```

```python
# plotting the graph
x = s1.index
y = s1.values
plt.bar(x,y)
```

<font color="red" size=3><b>Cumulative Distribution Function (cmf) of PMF</b></font>

```python
# calculating cumulative probability 
c1 = np.cumsum(s1)
c1
```

```python
# plotting the graph
x = c1.index
y = c1.values
plt.bar(x,y)
```
