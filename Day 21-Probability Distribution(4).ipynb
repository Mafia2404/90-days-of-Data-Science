<font color="blue" size=5><b>Probability Distribution Function- 4</b></font>

```python
import seaborn as sns
import numpy as np
import pandas as pd
import statsmodels.api as sm
import scipy.stats as stats
```

```python
df = sns.load_dataset('iris')
df.head()
```

<font color="red" size=3><b>Checking if a distribution is normal or not</b></font>

```python
sns.kdeplot(df['sepal_length'])
```

```python
# using statsmodel
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
# create QQ plot of 2 sets of datd
fig = sm.qqplot(df['sepal_length'], line='45', fit=True)
plt.title('QQ PLOT of normal distribution with normal fit')
plt.xlabel('Theoretical Quantile')
plt.ylabel('Sample Quantile')
plt.show()
```

<font color="red" size=3><b>Non Gaussian Distribution</b></font>

<font color="red" size=3><b>Uniform Distribution</b></font>

```python
x = np.random.uniform(low=0, high=1, size=1000)
plt.hist(x)
```

```python
# fit a uniform distribution to the data
p = stats.uniform.fit(x)
dist = stats.uniform(loc=p[0], scale=p[1])
```

```python
# QQ plot using uniform distribution
fig = sm.qqplot(x, dist=dist, line='45')
plt.title('QQ PLOT with uniform distribution with uniform fit')
plt.xlabel('Theoretical Quantile')
plt.ylabel('Sample Quantile')
plt.show()
```

<font color="red" size=3><b>Log Normal Distribution</b></font>

```python
x = np.random.lognormal(mean=0, sigma=1, size=1000)
plt.hist(x)
```

```python
p = stats.lognorm.fit(x)
dist = stats.lognorm(s=p[0], scale=np.exp(p[1]))
fig = sm.qqplot(x, dist=dist, line='45')
plt.title('QQ PLOT with Log Normal distribution with Log Normal fit')
plt.xlabel('Theoretical Quantile')
plt.ylabel('Sample Quantile')
plt.show()
```

<font color="red" size=3><b>Pareto Distribution</b></font>

```python
alpha = 2
xm = 1
x = stats.pareto.rvs(b=alpha, scale=xm, size=1000)
plt.hist(x)
```

```python
# fit a Pareto distribution to the data
params = stats.pareto.fit(x, floc=0)
dist = stats.pareto(b=params[0], scale=params[2])

# create QQ plot
fig = sm.qqplot(x, dist=dist, line='45')
plt.title('QQ PLOT with Pareto Distribution with Pareto Fit')
plt.xlabel('Theoretical Quantile')
plt.ylabel('Sample Quantile')
plt.show()
```
