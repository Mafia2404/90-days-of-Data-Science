```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
import numpy as np

# Create a concentric circle dataset
X, _ = make_circles(n_samples=500, factor=.5, noise=.03, random_state=4)

# Apply DBSCAN to the dataset
dbscan = DBSCAN(eps=0.1, min_samples=5)
clusters = dbscan.fit_predict(X)

# Plotting
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', marker='o')
plt.title("DBSCAN Clustering of Concentric Circles")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

```python
from sklearn.datasets import make_moons
X, _ = make_moons(n_samples=500, noise=0.05, random_state=4)
dbscan = DBSCAN(eps=0.2, min_samples=5)
clusters = dbscan.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', marker='o')
plt.title("DBSCAN Clustering of Moon-Shaped Data")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

```python
from sklearn.datasets import make_blobs
X, _ = make_blobs(n_samples=500, centers=3, cluster_std=0.5, random_state=4)
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', marker='o')
plt.title("DBSCAN Clustering of Blob Data")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

```python
import numpy as np
X = np.zeros((500, 2))
for i in range(500):
    r = i / 500 * 10
    theta = i * 4 * np.pi / 500
    X[i, 0] = r * np.cos(theta)
    X[i, 1] = r * np.sin(theta)
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', marker='o')
plt.title("DBSCAN Clustering of Spiral Data")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```
