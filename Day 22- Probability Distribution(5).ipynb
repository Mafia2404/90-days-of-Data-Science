<font color="blue" size=5><b>Central Limit Theorem</b></font>

```python
# set the parameters
num_samples = 1000
sample_size = 30
distribution_range = (0,1)
```

```python
# generate samples for uniform data
samples = np.random.uniform(distribution_range[0], distribution_range[1], (num_samples, sample_size))
print(samples.shape)
```

```python
# calculate sample mean
sample_mean = np.mean(samples, axis=1)
print(sample_mean)
```

```python
# plot the histogram
sns.distplot(sample_mean)
plt.title("Histogram of sample mean")
plt.xlabel('Sample mean')
plt.ylabel('Density')
plt.show()
```

```python
#  set the parameters
num_samples = 1000
sample_size = 50

# Poisson distribution
poisson_lambda = 5

# Gamma distribution parameters
gamma_shape = 2
gamma_scale = 1

# Binomial distribution parameter
binomial_n = 10
binomial_p = 0.5

```

```python
# generate samples from the distribution
poisson_samples = np.random.poisson(lam=poisson_lambda, size=(num_samples, sample_size))
gamma_samples = np.random.gamma(shape=gamma_shape, scale=gamma_scale, size=(num_samples, sample_size))
binomial_samples = np.random.binomial(n=binomial_n, p=binomial_p, size=(num_samples, sample_size))
```

```python
# calculate the sample mean
poisson_mean = np.mean(poisson_samples, axis=1)
gamma_mean = np.mean(gamma_samples, axis=1)
binomial_mean = np.mean(binomial_samples, axis=1)
```

```python
# plot the histogram of sample means
fig, axs = plt.subplots(3,1, figsize=(8,12))

axs[0].hist(poisson_mean, bins=20, density=True, edgecolor='black')
axs[0].set_title('Histogram of sample means of Poisson Distribution')
axs[0].set_xlabel('Sample mean')
axs[0].set_ylabel('Density')

axs[1].hist(gamma_mean, bins=20, density=True, edgecolor='black')
axs[1].set_title('Histogram of sample means ofGamma Distribution')
axs[1].set_xlabel('Sample mean')
axs[1].set_ylabel('Density')

axs[2].hist(binomial_mean, bins=20, density=True, edgecolor='black')
axs[2].set_title('Histogram of sample means of Binomial Distribution')
axs[2].set_xlabel('Sample mean')
axs[2].set_ylabel('Density')

fig.tight_layout()
plt.show()
```
