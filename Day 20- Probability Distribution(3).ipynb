```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

<font color="red" size=3><b>Understanding the use case of pdf</b></font>

```python
df = sns.load_dataset('iris')
df.head()
```

```python
sns.kdeplot(x=df['sepal_length'], hue=df['species'])
# not a good estimator of species
```

```python
sns.kdeplot(x=df['sepal_width'], hue=df['species'])
# not a good estimator of species
```

```python
sns.kdeplot(x=df['petal_length'], hue=df['species'])
# good estimator of species
```

```python
sns.kdeplot(x=df['petal_width'], hue=df['species'])
# good estimator of species
```

**Conclusion**: We'll keep petal length and petal width(Feature Selection)

<font color="red" size=3><b>Converting Normal Distribution to Standard Normal Distribution</b></font>

```python
df1.head()
```

```python
sns.kdeplot( x= df1['age'])
```

```python
# checking mean
u1 = df1['age'].mean()
u1
```

```python
# checking standard deviation
s1 = df1['age'].std()
s1
```

```python
x2 = (df1['age']-u1)/s1
```

```python
sns.kdeplot( x= x2)
```

```python
# checking mean
u2 = x2.mean()
u2 # close to zero
```

```python
# checking standard deviation
s2 = x2.std()
s2 # close to 1
```
